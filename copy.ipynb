```python
import pandas as pd
import numpy as numpy
import matplotlib.pyplot as plt 
import descartes
import geopandas as gpd 
from shapely.geometry import Point, Polygon
```

```python
df_stations = pd.read_csv("Stations.csv", names = ["Latitude", "Longitude", "Critical"])
df_connections = pd.read_csv("Connection_Times.csv", names = ["Station1", "Station2", "Times"])
```

```python
map_file = gpd.read_file("/Users/aleksander/Desktop/Netherlands_shapefile/nl_1km.shp")
fig,ax = plt.subplots(figsize= (15,15))
map_file.plot(ax=ax)
```

```python

geometry = [Point(xy) for xy in zip(df_stations["Longitude"],df_stations["Latitude"] )]

# fig,ax = plt.subplots(figsize = (15,15))
```

```python
geo_df = gpd.GeoDataFrame(df_stations, geometry = geometry)
fig,ax = plt.subplots(figsize = (15,15))
map_file.plot(ax=ax, alpha = 0.3, color = "grey")
geo_df[geo_df["Critical"]=="Critical"].plot(ax=ax, markersize = 20, color = "red", marker = "o", label = "Critical")
geo_df[geo_df["Critical"].isna()].plot(ax=ax, markersize = 20, color = "blue", marker = "^", label = "Non-Critical")
plt.legend(prop={'size':15})
```
